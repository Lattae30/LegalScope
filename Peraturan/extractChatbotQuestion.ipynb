```python
import pandas as pd
import json
import csv
```

```python
# Load the CSV file
file = "C:/Users/khalf/Code/LegalScope/LegalScopeChatbot/chatbotQuestion.csv"
df = pd.read_csv(file)

# Display the first few rows and column names for debugging
print("Columns in the CSV file:", df.columns)
print(df.head())

# Ensure correct column names
required_columns = ['No. ', 'Pertanyaan', 'Jawaban']
if not all(col in df.columns for col in required_columns):
    raise ValueError(f"CSV file must contain the following columns: {required_columns}")

# Extract relevant columns and drop rows with missing values
df_extracted = df[required_columns].dropna()

# Rename the 'No' column to 'qna X'
df_extracted['No. '] = df_extracted['No. '].apply(lambda x: f'qna {str(x).strip()}')

# Rename columns
df_extracted = df_extracted.rename(columns={'No. ': 'no', 'Pertanyaan': 'pertanyaan', 'Jawaban': 'jawaban'})

# Display the extracted data for debugging
print("Extracted Data:")
print(df_extracted.head())
print(f"Total questions extracted: {len(df_extracted)}")

# Convert to a list of dictionaries
qna_list = df_extracted.to_dict(orient='records')

# Create the intents structure
intents = [
    {
        "tag": qna['no'],
        "patterns": [qna['pertanyaan']],
        "responses": [qna['jawaban']],
        "context": [""]
    }
    for qna in qna_list
]

# Create the final JSON structure
data = {
    "intents": intents
}

# Save to JSON
output_file = 'C:/Users/khalf/Code/LegalScope/LegalScopeChatbot/intents.json'
with open(output_file, 'w', encoding='utf-8') as json_file:
    json.dump(data, json_file, indent=2, ensure_ascii=False)

print("Data has been successfully extracted and saved to intents.json")
```
